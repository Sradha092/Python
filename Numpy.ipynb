### Copy and View
### The main difference between copy and view of an array is that the copy creates a new array and view is just the view of original array
### Similar to shallowcopy and deepcopy in list
### Whenever there is a nested list,shallowcopy fails, hence the need of deepcopy

```python
arr=np.array([1,2,3,4,5])
print(arr,id(arr))
```

```python
x=arr.copy()
print(x,id(x))    #stored in different location
```

```python
x[0]=42
```

```python
x
```

```python
arr      #Changing the copy doesn't change the original array
```

```python
arr=np.array([1,2,3,4,5])
print(arr,id(arr))

x=arr.view()
print(x,id(x))  #Stored in different location
x[0]=42
print(x)
print(arr)      #Changing the view object changes the original array
```
