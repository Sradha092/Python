```python
import seaborn as sns
```

```python
#Seaborn has its own data server-seaborn load dataset-google

tips=sns.load_dataset('tips')
```

```python
tips
```

## Relational plot

```python
#Only two numerical data, others categorical data
sns.relplot(x='total_bill',y='tip',data=tips)
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',data=tips)  #categorize based on color
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='time',data=tips) #style-categorizing based on different icons
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='time',size='size',data=tips) #size-categorize based on size of different icons
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='time',col='time',data=tips) 
```

## Bar Plot

```python
#One should be categorical and one should be numerical
sns.barplot(x='time',y='total_bill',data=tips)
```

## Categorical plot

```python
sns.catplot(x='day',y='total_bill',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',data=tips,jitter=False) #jitter-to display in a straight line
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='swarm')
```

```python
#boxplot-the data which is not following the trend-outlier
sns.catplot(x='day',y='total_bill',data=tips,kind='box')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='violin')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='boxen')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='point')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
sns.catplot(x='day',data=tips,kind='count')
```
